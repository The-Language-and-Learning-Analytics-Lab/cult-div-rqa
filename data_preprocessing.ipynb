```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./data/CPS_Chat_Data_F19_20200129.xlsx - uci_data_collection_completed_a.csv')
```

```python
data = data[data.event_name == 'chat']
```

```python
data.columns
```

```python
coded_data = pd.read_csv('./data/Master Coded Data.xlsx - Chats.csv')
```

```python
coded_data.columns
```

```python
coded_data = coded_data.rename(columns={'Person-ID (mellon-ID)' : 'mellon_id'})
data = data.rename(columns={'event_by' : 'mellon_id', 'team_id (mellon_id)':'team_id'})
```

```python
coded_data.head()
```

```python
merged_data = data.merge(coded_data[['mellon_id', 'event_result', 'event_start_time', 'event_end_time', 'Primary code ']], on = ['mellon_id', 'event_result', 'event_start_time', 'event_end_time'], how = 'left')
```

```python
merged_data[merged_data.isna().any(axis=1)]
```

```python
merged_data_clean = merged_data.dropna()
```

```python
scores = pd.read_csv('./data/CPS_scores.csv')
```

```python
merged_data_clean_scores = merged_data_clean.merge(scores[['mellon_id','revise_score']], on=['mellon_id'], how = 'left')
```

```python
merged_data_clean_scores.head()
```

```python
d1 = merged_data_clean_scores.groupby(by=['session_id', 'mellon_id'], as_index=False).agg({'revise_score': 'mean'})
```

```python
d2 = d1.groupby(by=['session_id'], as_index=False).agg({'revise_score': 'mean'})
```

```python
d2
```

```python
merged_data_clean_scores
```

```python
d3 = merged_data_clean_scores[['session_id', 'mellon_id', 'player_id', 'event_result', 'task', 'Primary code ', 'event_start_time']].merge(d2, on = ['session_id'])
```

```python
d3.head()
```

```python
mean = np.mean(d3['revise_score'])
high_perf = (d3['revise_score'] > mean).astype(int)
```

```python
d3['high_performance'] = high_perf
```

```python
d3.head()
```

```python
groups = d3.groupby(by=['session_id'])['mellon_id'].unique()
```

```python
background = pd.read_csv('./data/Mellondemographic_Student_20200205.csv')
```

```python
background.columns
```

```python
female = []
fm = []
non_int = ['wendy6', 'Ratlife2001', 'Isabella', 'juarezm', 'jennajyee', 'Santa@na2019', 'ucinetid', 'zotzot', 'C@ss3tte']
for i in d3.session_id:
    s = 0
    members = groups[i]
    for j in members:
        if j in non_int:
            continue
        g = background[background.mellon_id == int(j)]['gender'].any()
        if g == 'F':
            s += 1
    female.append(s)
    fm.append(int(s>(len(members)/2)))
```

```python
d3['female_members'] = female
d3['female_maj'] = fm
```

```python
d3
```

```python
d3.mellon_id = d3.mellon_id.astype(str)
background.mellon_id = background.mellon_id.astype(str)
```

```python
d3.merge(background[['mellon_id', 'gender', 'ethnicity', 'application_status', 'first_generation', 'urm', 'first_language']], on = 'mellon_id', how = 'left').to_csv('./data/final_final.csv')
```
